```python
year = [2018,2017,2016,2015,2014,2013]
sales = [108203, 100904, 94595, 83176, 78812, 74754 ]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score


x = year
y = sales
p1 = np.poly1d(np.polyfit(x, y, 1))
p2 = np.poly1d(np.polyfit(x, y, 3))
p3 = np.poly1d(np.polyfit(x, np.log(y), 2))

xpred = [2019,2020]
t = np.linspace(2013, 2020, 200)

fig, axes = plt.subplots(1,3, figsize=(14, 3))

axes[0].plot(x, y, 'o', t, p1(t), '--')
axes[0].plot(xpred,p1(xpred),'r*')
#r1 = r2_score(y,p1(x))
axes[0].set_title('Linear prediction: R2 score:{:2f}'.format(r2_score(y,p1(x))))

axes[1].plot(x, y, 'o', t, p2(t), '--')
axes[1].plot(xpred,p2(xpred),'r*')
axes[1].set_title('Polynomial fit: R2 score:{:2f}'.format(r2_score(y,p2(x))))

axes[2].plot(x, y, 'o', t, np.exp(p3(t)), '--')
axes[2].plot(xpred, np.exp(p3(xpred)),'r*')
axes[2].set_title('Logistic regression: R2 score:{:2f}'.format(r2_score(y,np.exp(p3(x)))));

print(p1(xpred),p2(xpred),np.exp(p3(xpred)))
```
